### Data Preprocessing

```python
#Encoding
le=LabelEncoder()
```

```python
labeling_features=salary_train.columns
labeling_features=labeling_features.drop(['age','educationno','capitalgain','capitalloss','hoursperweek'])
```

```python
for i in labeling_features:
    salary_train[i]=le.fit_transform(salary_train[i])
salary_train
```

```python
for i in labeling_features:
    salary_test[i]=le.fit_transform(salary_test[i])
salary_test
```

```python
#Standardize the data
#Training data
scaler=MinMaxScaler()
salary_train_scaled=scaler.fit_transform(salary_train)
salary_train_scaled=pd.DataFrame(salary_train_scaled,columns=salary_train.columns)
salary_train_scaled
```

```python
#Testing Data
salary_test_scaled=scaler.fit_transform(salary_test)
salary_test_scaled=pd.DataFrame(salary_test_scaled,columns=salary_test.columns)
salary_test_scaled
```

### Model Building

```python
x_train=salary_train_scaled.drop('Salary',axis=1)
y_train=salary_train_scaled[['Salary']]
```

```python
x_test=salary_test_scaled.drop('Salary',axis=1)
y_test=salary_test_scaled[['Salary']]
```

### Model Training | Testing | Evaluation

#### Using Gaussian Naive Bayes

```python
classifier=GaussianNB()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy Score:',accuracy_score(y_test,y_pred))
print('Confusion Matrix :\n',confusion_matrix(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.show()
```

#### Using Multinomial Naive Bayes

```python
classifier=MultinomialNB()
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy Score:',accuracy_score(y_test,y_pred))
print('Confusion Matrix :\n',confusion_matrix(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True)
plt.show()
```

#### From the above result, Gaussian Model gives better result with 79% accuracy
